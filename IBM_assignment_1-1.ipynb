```python


import re

df = pd.read_csv('House Price India.csv')

data = pd.DataFrame(df)
```

UNIVARIATE VISUALIZATION OF THE DATASET

Analysis the data

```python
# View first 10 rows of dataset
print(data.head(10))

# Check data types of columns
print(data.dtypes)

# Check for missing values
print(data.isnull().sum())
```

```python
# Select the variable of interest
var_of_interest = 'Price'
```

```python
# Compute descriptive statistics for the variable of interest
print(data[var_of_interest].describe())
```

Histogram of the variable

```python
# Create a histogram of the variable of interest
plt.hist(data[var_of_interest], bins=20)
plt.xlabel(var_of_interest)
plt.ylabel('Frequency')
plt.title('Histogram of {}'.format(var_of_interest))
plt.show()
```

BOX PLOT OF THE VARIABLE

```python
# Create a boxplot of the variable of interest
plt.boxplot(data[var_of_interest])
plt.ylabel(var_of_interest)
plt.title('Boxplot of {}'.format(var_of_interest))
plt.show()
```

DENSITY PLOT OF THE VARIABLE

```python
# Create a density plot of the variable of interest
data[var_of_interest].plot(kind='density')
plt.xlabel(var_of_interest)
plt.title('Density plot of {}'.format(var_of_interest))
plt.show()
```

```python
# Compute the skewness and kurtosis of the variable of interest
print('Skewness: {:.2f}'.format(data[var_of_interest].skew()))
print('Kurtosis: {:.2f}'.format(data[var_of_interest].kurt()))
```
